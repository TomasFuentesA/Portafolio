```python
import pandas as pd

vehicles = pd.read_csv('vehicles.csv')
vehicles.info()
```

```python
vehicles.head()
```

<h1 align=center>Simple Aggregations</h1>

`describe()` method returns a statistical summary for each of the columns in a DataFrame 

```python
vehicles[['model']].describe()
```

```python
vehicles[['co2emissions']].describe()
```

```python
vehicles[['model']].value_counts()
```

```python
vehicles[['model']].value_counts(normalize= True)
```

```python
vehicles[['co2emissions']].mean()
```

<h1 align='center'> Group-level Aggregations </h1>

```python
vehicles.groupby('model')[['co2emissions']].mean()
```

```python
vehicles.groupby('model')[['co2emissions']].mean().sort_values(by= 'co2emissions')
```

```python
vehicles.groupby('model')[['co2emissions']].agg(['mean', 'median','min','max'])
```
